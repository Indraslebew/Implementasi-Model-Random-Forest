```python
import pandas as pd
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
df
```

```python
#mengubah data yang bertipe string menjadi angka
df ['Name'] = df['Name'].astype('category')
df ['Name'] = df['Name'].cat.codes

df ['Sex'] = df['Sex'].astype('category')
df ['Sex'] = df['Sex'].cat.codes

df['Ticket'] =df['Ticket'].astype('category')
df['Ticket'] =df['Ticket'].cat.codes

df ['Cabin'] = df['Cabin'].astype('category')
df ['Cabin'] = df['Cabin'].cat.codes

df['Embarked'] = df['Embarked'].astype('category')
df['Embarked'] = df['Embarked'].cat.codes

df
```

```python
#pemeriksaan baris kosong(missing)
df.isnull().sum()
```

```python
df = df.dropna()
df
```

```python
x = df.drop(columns ='Embarked')
y = df['Embarked']
y
```

```python
#membagi data menjadi data training dan testing(splitting data)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#melakukan feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion = "gini", max_depth = 8, min_samples_split = 10, random_state = 5)
```

```python
clf.fit(x_train, y_train)
```

```python
#PREDIKSI DATA
y_pred = clf.predict(x_test)
y_pred
```

```python
y_pred_df = pd.DataFrame({'Nilai aktual': y_test, 'Nilai Prediksi': y_pred})
y_pred_df.head(10)
```

```python
clf.predict(x_test)
```

```python
akurasi_training = clf.score(x_train,y_train)
print('Akurasi Training = ', akurasi_training)
```

```python
akurasi_testing = clf.score(x_test,y_test)
print('Akurasi Testing = ', akurasi_testing)
```
